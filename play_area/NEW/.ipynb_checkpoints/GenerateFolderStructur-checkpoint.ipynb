```python
import os
import shutil
import time
```

```python
class GenerateFolderStructur():
    """
    GeneratedData
    |
    -> InitalImages
    |  |-> A-Z;0-9;/;-;.;:
    |
    -> TempFolderRename
    |  
    |
    -> DistortedImages
    |  |-> A-Z;0-9;/;-;.;:
    |
    -> TailoredImages
    |  |-> A-Z;0-9;/;-;.;:
    |
    -> BackgroundImages
    |
    |
    -> SegmentedImages
    |  
    |
    -> FinalImages
    |
    |
    -> XML-Files
    |
    |
    -> TempFolderDistortion
    |
    |
    -> data    
    
    GeneratedData => Ordner
    InitalImages => Zeichen/Bilder werden hier abgelegt
    TempFolderRename => Dient nur als Zwischenspeicher zur Umbennenung der Bilder
    DistortedImages => Verzerrtes Bild
    TailoredImages => Zugeschnittenes verzerrtes Bild 
    BackgroundImages => Hintergrundbild
    SegmentedImages => Segmentiertes Bild (Hintergrund, Vordergrund)
    FinalImages => BackgroundImages + Mehrere TailoredImages
    XML-Files => XML-File mit Bounding boxes
    TempFolderDistortion => Bilder werden hier in verschoben und dann verzerrt, dient nur als Zwischenspeicher 
    data => Yolov3 Struktur für training
    
    Die Klasse kann mit folgenden Parameter initalisiert werden:
    Ordnerpfad: 'C:/Users/User/Desktop/test/', 
    Klassenordner: ['A-Z','0','-','u','10-15','rdm_name','a-c']
    """
    
    dictSigns = {'-':'Dash','.':'Dot','/':'ForwardSlash',':':'Colon'}
    all_characters = ('ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz')
    numbers = ('0123456789')
    characters_folder = []
    
    def __init__(self, path,  characters):
        self._path = path + '/GeneratedData/'
        self._characters = characters
        self._initalImages_path = path + '/GeneratedData/InitalImages/'
        self._tempFolderRename_path = path + '/GeneratedData/TempFolderRename/'
        self._distortedImages_path = path + '/GeneratedData/DistortedImages/'
        self._tailoredImages_path = path + '/GeneratedData/TailoredImages/'
        self._backgroundImages_path = path + '/GeneratedData/BackgroundImages/'
        self._segmentedImages_path = path + '/GeneratedData/SegmentedImages/'
        self._finalImages_path = path + '/GeneratedData/FinalImages/'
        self._xml_Files_path = path + '/GeneratedData/XML_Files/'
        self._tempFolderDistortion_path = path + '/GeneratedData/TempFolderDistortion/'
        self._data_path = path + '/GeneratedData/data/'
        self.set_folder_structur()
        
        path, dirs, files = next(os.walk(self._initalImages_path))
        with open(self._data_path + '/' + 'obj.names', "w+") as file:
            [file.write(name + '\n') for name in dirs]
        
        
    def set_characters_folder(self):
        for i in range(len(self._characters)):
### For -./: and single characterss
            if len(self._characters[i]) == 1:
#                 print('singleShit')
                if self._characters[i] in type(self).dictSigns:
                    type(self).characters_folder.append(type(self).dictSigns[self._characters[i]]) 
                    continue
#                 if self._characters[i] == '-':
#                     type(self).characters_folder.append('Dash') ### DASH
#                     continue
#                 elif self._characters[i] == '.':
#                     type(self).characters_folder.append('Dot') ### DOT
#                     continue
#                 elif self._characters[i] == '/':
#                     type(self).characters_folder.append('ForwardSlash') ### FORWARD_SLASH
#                     continue
#                 elif self._characters[i] == ':':
#                     type(self).characters_folder.append('Colon') ### COLON
#                     continue
                else:
                    if self._characters[i].isupper() == True:
                        type(self).characters_folder.append(self._characters[i] + '_upperCase')
                        continue
                    else:
                        type(self).characters_folder.append(self._characters[i] + '_lowerCase')
                        continue
### for Names    
            if len(self._characters[i]) > 1 and self._characters[i].find('-') == -1:
#                 print('forNames')
                type(self).characters_folder.append(self._characters[i])
                
### for A-Z or a-z   
            if type(self).all_characters.find(self._characters[i][0]) != -1 and self._characters[i].find('-') != -1:
#                 print('A-Z  a-z')
                for j in range(type(self).all_characters.find(self._characters[i][0]), type(self).all_characters.find(self._characters[i][2])+1):
#                     type(self).characters_folder.append(type(self).all_characters[j])
                    if j <= 25:
                        type(self).characters_folder.append(type(self).all_characters[j] + '_upperCase')
                    else:
                        type(self).characters_folder.append(type(self).all_characters[j] + '_lowerCase')
                    
### for 1023-1026
            elif self._characters[i].find('-') != -1 and type(self).numbers.find(self._characters[i][0]) != -1: 
#                 print('1023-1026')
                tmp_0 = ''
                tmp_1 = ''
                for u in range(0, self._characters[i].find('-')):
                    tmp_0 = tmp_0 + self._characters[i][u]
                for v in range(self._characters[i].find('-')+1, len(self._characters[i])):
                    tmp_1 = tmp_1 + self._characters[i][v]
                for k in range(int(tmp_0), int(tmp_1)+1):
                    type(self).characters_folder.append(str(k))
#                 print(tmp_0,tmp_1)    
            
#         print(type(self).characters_folder)    
        for m in range(len(type(self).characters_folder)):
#             print(self._initalImages_path + type(self).characters_folder[m])
            
#             os.mkdir(self._initalImages_path + type(self).characters_folder[m] + '_' + f'{m:06}')                                              
#             os.mkdir(self._tempFolderRename_path + type(self).characters_folder[m] + '_' + f'{m:06}')
#             os.mkdir(self._distortedImages_path + type(self).characters_folder[m] + '_' + f'{m:06}')
#             os.mkdir(self._tailoredImages_path + type(self).characters_folder[m] + '_' + f'{m:06}')
            os.mkdir(self._initalImages_path + type(self).characters_folder[m])                                              
#             os.mkdir(self._tempFolderRename_path + type(self).characters_folder[m])
            os.mkdir(self._distortedImages_path + type(self).characters_folder[m])
            os.mkdir(self._tailoredImages_path + type(self).characters_folder[m])
            
        
    def set_folder_structur(self):
        os.mkdir(self._path)
        os.mkdir(self._initalImages_path)
        os.mkdir(self._tempFolderRename_path)
        os.mkdir(self._distortedImages_path)
        os.mkdir(self._tailoredImages_path)
        os.mkdir(self._backgroundImages_path)
        os.mkdir(self._segmentedImages_path)
        os.mkdir(self._finalImages_path)
        os.mkdir(self._xml_Files_path)
        os.mkdir(self._tempFolderDistortion_path)
        os.mkdir(self._data_path)
        if len(self._characters) > 0:
            self.set_characters_folder()
        
    
```
